#### hurm, i just realised that each of the datasets is having the same issues:
1. column name
2. details in name's column having aliases/@.
3. unexpected index.

I think better to create a function to extract the data for all  four datasets


```python
# Craet a function that can be used over and over again

def extract_clean(df) :
     #renaming columns
    df.rename(columns = {
        'NAME':'name',
    'FOLLOWERS':'followers',
    'ER':'engagement_rate',
    'TOPIC OF INFLUENCE':'topic',
    'POTENTIAL REACH':'potential_reach'}, inplace=True)
    
    # clean the name column by removing '@'
    df["name"] = df["name"].str.split("@").str[0]
    
    #selecting columns
    df = df[["name","followers","engagement_rate","topic","potential_reach"]]
    return df
    
```

```python
tiktok = pd.read_csv("tiktok_data_malaysia.csv")
df_tt = extract_clean(tiktok)
df_tt.head()
```
